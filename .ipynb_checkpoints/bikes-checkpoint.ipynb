```python
import pandas as pd
import seaborn as sns
import numpy as np
# might not be necessary in notebook because it's already handled
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/hour.csv')
```

```python
# 100 first records
df.head(100)
```

```python
# 10 last records
df.tail(10)
```

```python
# 10 random records
df.sample(10)
```

```python
# descripctive statistics of dataframe
df.describe()
```

```python
# variant of rented bikes number related to the hour of the day
sns.barplot(
    # whole dataframe
    data=df,
    # feature of dataframe on X axis (hour of the day)
    x='hr',
    # feature of dataframe on Y axis (count of bikes)
    y='cnt'
)
```

```python
# features of dataframe
df.columns
```

```python
# count of render bikes related to many features
for i in ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']:
    sns.barplot(data=df, x=i, y='cnt')
    # because it would print all graphs one above the other
    plt.show()
```

```python
for i in ['temp', 'atemp', 'hum', 'windspeed']:
    sns.scatterplot(data=df, x=i, y='cnt')
    plt.show()
```

```python
sns.jointplot(
    data=df,
    x='temp',
    y='cnt',
    # kernel density extimation => ex. geographical chart
    kind='kde'
)
```

```python
# cross features graphs, how features are related
sns.pairplot(
    data=df[ ['temp', 'hum', 'windspeed', 'cnt'] ]
)
```

```python
# combine workingdays to the hour with different colors
sns.barplot(
    data=df,
    x='hr',
    y='cnt',
    hue='workingday'
)
```

```python
sns.relplot(
    data=df,
    x='hr',
    y='cnt',
    # display seasons with different colors
    hue='season',
    # 2 graphs columns with working day 0 and working day 1
    col='workingday',
    # default is scatterplot
    kind='line',
    # scale sizes basing on count
    size='cnt'
)
```

```python
sns.relplot(
    data=df,
    x='temp',
    y='hum',
    hue='workingday',
    col='yr',
    row='weathersit'
)
```

```python
sns.boxplot(
    data=df,
    x='season',
    y='cnt',
    hue='holiday'
)
```

```python
sns.violinplot(
    data=df,
    x='season',
    y='cnt',
    hue='holiday'
)
```
